```python
import requests
import re

stock_id = {}
for i in alphabet_list:
    url = f'https://www.moneycontrol.com/india/stockpricequote/{i}'
    r = requests.get(url)
    html_content = r.text

    pattern = re.compile(r'href="(https://www\.moneycontrol\.com[^"]+)" class="bl_12">([^<]+)</a>')


    matches = re.findall(pattern, html_content)


    for match in matches:
        link, stock_name = match
        stock_id[stock_name] = link.split("/")[-1]

stock_id



```

```python
import csv


header = ['stock_id', 'mar_2023', 'mar_2022', 'mar_2021', 'mar_2020', 'mar_2019']
rows = []

for stock_ids, values in final_stocks_dict.items():
    row = [stock_ids] + values
    rows.append(row)


csv_file_path = 'output.csv'
with open(csv_file_path, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    

    csv_writer.writerow(header)

    csv_writer.writerows(rows)

print(f'CSV file "{csv_file_path}" has been created.')
```

```python
def adjacent_element_product(array):
    l = list(zip(array, array[1:]))
    s = [a*b for a, b in zip(array, array[1:])]
    return max( a*b for a, b in zip(array, array[1:]) ),s,l

z = [1,2,3,4,5]
print(adjacent_element_product(z))
```
