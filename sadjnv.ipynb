```python
import requests
import re

stock_id = {}
for i in alphabet_list:
    url = f'https://www.moneycontrol.com/india/stockpricequote/{i}'
    r = requests.get(url)
    html_content = r.text

    pattern = re.compile(r'href="(https://www\.moneycontrol\.com[^"]+)" class="bl_12">([^<]+)</a>')


    matches = re.findall(pattern, html_content)


    for match in matches:
        link, stock_name = match
        stock_id[stock_name] = link.split("/")[-1]

stock_id



```

```python
for i in stock_id:
    print(stock_id[i])
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

def fetch_stock_data(stock):
    try:
        url = f'https://www.moneycontrol.com/financials/results/yearly/{stock_id[stock]}'
        r = requests.get(url)
        html_content = r.text
        soup = BeautifulSoup(html_content, 'html.parser')
        stock_name = soup.find_all("td")[1].text.strip()
        return stock, stock_name
    except Exception as e:
        return None


with ThreadPoolExecutor(max_workers=5) as executor:
    results = list(executor.map(fetch_stock_data, stock_id.keys()))

filtered_results = [result for result in results if result is not None]


df = pd.DataFrame(filtered_results, columns=['Stock', 'Symbol'])

print(df)


        
        

```
